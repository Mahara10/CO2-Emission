```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Region', y='Metric Tons Per Capita', palette='viridis')
plt.title('Distribution of CO₂ Emissions Per Capita by Region')
plt.xlabel('Region')
plt.ylabel('Metric Tons Per Capita')
plt.xticks(rotation=45)
plt.show()
```

```python
growth_rate = (
    df.groupby(['Country'])
    .apply(lambda x: (x['Kilotons of Co2'].iloc[-1] - x['Kilotons of Co2'].iloc[0]) / x['Kilotons of Co2'].iloc[0] * 100)
    .reset_index(name='Growth Rate (%)')
    .sort_values('Growth Rate (%)', ascending=False)
    .head(10)
)

plt.figure(figsize=(12, 6))
sns.barplot(data=growth_rate, x='Growth Rate (%)', y='Country', palette='coolwarm')
plt.title('Top 10 Countries with the Highest CO₂ Growth Rate')
plt.xlabel('Growth Rate (%)')
plt.ylabel('Country')
plt.show()
```

```python
heatmap_data = df.groupby([df['Date'].dt.year, 'Region'])['Kilotons of Co2'].sum().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".0f", linewidths=.5)
plt.title('Heatmap of CO₂ Emissions by Region and Year')
plt.xlabel('Region')
plt.ylabel('Year')
plt.show()
```

```python
total_emissions = df.groupby(df['Date'].dt.year)['Kilotons of Co2'].sum().reset_index(name='Global Total')
top_emitters = (
    df.groupby(['Date', 'Country'])['Kilotons of Co2'].sum().reset_index()
    .sort_values(by=['Date', 'Kilotons of Co2'], ascending=[True, False])
    .groupby('Date')
    .head(10)
    .groupby('Date')['Kilotons of Co2'].sum()
    .reset_index(name='Top 10 Emitters Total')
)

total_emissions['Date'] = pd.to_datetime(total_emissions['Date'], format='%Y')


merged = pd.merge(total_emissions, top_emitters, on='Date')
merged['Contribution (%)'] = (merged['Top 10 Emitters Total'] / merged['Global Total']) * 100

plt.figure(figsize=(10, 6))
sns.lineplot(data=merged, x='Date', y='Contribution (%)', marker='o')
plt.title('Top 10 Emitters Contribution to Global CO₂ Emissions')
plt.ylabel('Contribution (%)')
plt.xlabel('Year')
plt.show()
```
